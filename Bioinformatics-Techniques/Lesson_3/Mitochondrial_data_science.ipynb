```python
def get_data_files(directory):
    for root, directories, files in walk(directory):
        for file in files:
            current_path = path.join(root, file)
            yield current_path
    
```

```python
def get_data_from_gb(path):
    sequence = ''
    collect_sequence = 0
    with open(path, "rt") as gb:
        for line in gb:
            if line.startswith('ACCESSION'):
                uniprot_id = line[13:].strip()
                continue
            if line.startswith('ORIGIN'):
                collect_sequence = 1
                continue
            if collect_sequence:
                sequence += line[11:]
        return uniprot_id, sequence
```

```python
df = pd.DataFrame(columns=['ID', 'sequence', 'organism'])
```

```python
for file in get_data_files('./python-in-data-science-data/genomes/'):
    name = path.basename(file).split('.')[0]
    data = get_data_from_gb(file)
    df = df.append({'ID': data[0],
                    'sequence': data[1].upper(),
                    'organism': name}, ignore_index=1)
```

```python
df['sequence_length'] = df['sequence'].apply(lambda x: len(x))
```

```python
bases = ['a', 'c', 'g', 't']
for base in bases:
    df[base] = df['sequence'].apply(lambda x: len([a for a in x if a==base.upper()]))
```

```python
df['gc'] = df['g']+df['c']
```

```python
df['gc%'] = df['gc']/df['sequence_length']
```

```python
organisms_list = df['organism'].unique()
```

```python
for organism in organisms_list:
    plt.bar(x=organism, height=df[df['organism']== organism]['gc'])
plt.title("Zad 1: Amount of CG")
plt.show()
```

```python
for base in bases:
    for organism in organisms_list:
        plt.bar(x=organism, height=df[df['organism']==organism][base])
    plt.title("Zad 2: Base {}".format(base.upper()))
    plt.show()
```

```python
for organism in organisms_list:
    plt.bar(x=organism, height=df[df['organism']== organism]['gc%'])
plt.title("Zad 3: GC content")
plt.show()
```
